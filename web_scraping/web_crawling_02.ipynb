## title 출력

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://finance.naver.com/news/news_list.nhn?mode=LSS2D&section_id=101&section_id2=258&date=20210110'
res = requests.get('https://finance.naver.com/news/news_list.nhn?mode=LSS2D&section_id=101&section_id2=258')
```

```python
def nextpage(date,endpage):
    for page in range(int(endpage)):
        res = requests.get('https://finance.naver.com/news/news_list.nhn?mode=LSS2D&section_id=101&section_id2=258&date={}{}'.format(str(date), '&page=' + str(page+1)))
        soup = BeautifulSoup(res.text, 'lxml')
        title = soup.find_all('dd', class_='articleSubject')

        for i in title:
            print(i.get_text())
```

```python
soup = BeautifulSoup(res.text, 'lxml')
main = soup.find('title')
date = url.split('&date=')[1][:8]
endpage = soup.find('td', class_="pgRR")
endpage = endpage.a['href'].split('=')[-1]
print(main.string)
print(endpage)
```

```python
nextpage(date, endpage)
```
