```python
import numpy as np
import librosa, librosa.display 
import matplotlib.pyplot as plt
from pydub import AudioSegment
import os
```

```python
path = './source'
filename = '20210106.m4a'

file = os.path.join(path, filename)
print(file)
```

```python
sound = AudioSegment.from_file(file)
sound = sound.set_channels(1)

dstfilename = filename.split('.')[0]
dstfile = os.path.join(path, dstfilename+'.wav')
print(dstfile)

sound.export(dstfile, format="wav")
```

```python
sig, sr = librosa.load(dstfile, sr=22050)

print(sig,sig.shape)
```

```python
FIG_SIZE = (15,10)
plt.figure(figsize=FIG_SIZE)
librosa.display.waveplot(sig, sr, alpha=0.5)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform")
```
