```python
import random

# 날짜 생성함수
def givemedate():
    choice = random.randint(1,6)
    ref = [1, 0]
    date_list = []
    
    # 년월일, use or not use 요일
    if choice == 1:
        sel_year = random.choice(year)
        sel_month = random.choice(month)
        sel_day = random.choice(day)
        date_list.extend([sel_year,sel_month,sel_day])
        select = random.choice(ref)
        if select:
            dow = random.choice(week_day)
            date_list.append(dow)
    # 월일, use or not use 요일
    elif choice == 2:
        sel_month = random.choice(month)
        sel_day = random.choice(day)
        date_list.extend([sel_month,sel_day])
        select = random.choice(ref)
        if select:
            dow = random.choice(week_day)
            date_list.append(dow)
    # 일, use or not use 요일
    elif choice == 3:
        sel_day = random.choice(day)
        date_list.append(sel_day)
        select = random.choice(ref)
        if select:
            dow = random.choice(week_day)
            date_list.append(dow)
    # 첫째주, 둘째주 출력, use or not use 요일 or 월
    elif choice == 4:
        week_num = random.choice(num_week)
        date_list.append(week_num)
        
        select = random.choice(ref)
        if select:
            dow = random.choice(week_day)
            date_list.append(dow)
            
        if not(select):
            sel_month = random.choice(month)
            date_list.insert(0,sel_month)
            
    elif choice == 5:
        known_d = knowndate + day
        sel_day = random.choice(known_d)
        date_list.append(sel_day)
        if sel_day.endswith('개월' or '일'):
            select = random.choice(ref)
            if select:
                date_list.pop()
                date_list.append(sel_day+'후')
            else:
                date_list.pop()
                date_list.append(sel_day+'뒤')
                
    elif choice == 6:
        sel_day = random.choice(unknowndate)
        date_list.append(sel_day)
        
        if sel_day.endswith('달'):
            week_num = random.choice(num_week)
            date_list.append(week_num)
            
        elif sel_day.endswith('년'):
            sel_month = random.choice(month)
            sel_day = random.choice(day)
            date_list.extend([sel_month,sel_day])
            

        dow = random.choice(week_day)
        date_list.append(dow)
            
    return date_list


def givemetime():
    ref = [1, 0]
    minute_list = [minute]
    time_list = []
    
    # 시간 리스트 합치기
    hour = hour_12 + hour_24 + other_time
    # 시간출력 
    # 12시간제라면 use or not use 오전오후
    # + 시간 출력(00분, ~분전, ~분후)
    # 정각 / 정오라면 ~분전,~분후만출력

    sel_hour = random.choice(hour)
    time_list.append(sel_hour)

    if sel_hour in hour_12:
        select = random.choice(ref)
        if select:
            time_noun = random.choice(hour_12_m)
            time_list.insert(0,time_noun)
    # ex) 1시간 10분전
    elif sel_hour in other_time:
        select = random.choice(ref)
        if select:
            ago = random.choice(time_a)
            time_list.append(ago)
        else:
            later = random.choice(time_b)
            time_list.append(later)
        return time_list

    sel_min = random.choice(minute)
    time_list.append(sel_min)
    
    select = random.choice(ref)
    if select:
        aff = random.choice(affix)
        time_list.pop()
        time_list.append(sel_min+aff)
        
    return time_list

# 날짜가 오늘, 금일 일 때만 사용 가능
def few_minute():
    ref = [1, 0]
    hour = hour_12 + hour_24
    # ~분후 일정잡기
    time_list = []
    
    select = random.choice(ref)
    if select:
        ago = random.choice(hour)
        time_list.append(ago+'간')
        
    select = random.choice(ref)
    if select:
        sel_min = random.choice(time_a)
        time_list.append(sel_min)
    else:
        if len(time_list) != 0:
            time_list.append('후')
    
    select = random.choice(ref)
    if len(time_list) != 0:
        if select:
            aff = random.choice(affix)
            time_list.append(aff)
    return time_list
        
def givemeending():
    #stem1, object_p
    ref = [1,0,0,1,1,0]
    
    ending_list = []
    stem = random.choice(stem1)
    object_do = random.choice(object_p)
    ending_list.append(stem)
    
    if stem == '울려':
        ending_list.pop()
        ending_list.append('일정 '+stem)
        
        select1 = random.randint(1,3)
        if select1 == 1:
            ending_list.pop()
            ending_list.append('일정 '+stem+' '+object_do)
        elif select1 == 2:
            ending_list.pop()
            ending_list.append('일정'+' '+object_do+' '+stem)
        elif select1 == 3:
            ending_list.pop()
            ending_list.append(object_do+' '+'일정'+' '+stem)
            
    elif stem == '때려줘':
        ending_list.pop()
        ending_list.append('일정 '+stem)
        
        select1 = random.randint(1,3)
        if select1 == 1:
            ending_list.pop()
            ending_list.append('일정 '+stem+' '+object_do)
        elif select1 == 2:
            ending_list.pop()
            ending_list.append('일정'+' '+object_do+' '+stem)
        elif select1 == 3:
            ending_list.pop()
            ending_list.append(object_do+' '+'일정'+' '+stem)
            
    elif stem == '짜줘':
        ending_list.pop()
        ending_list.append('일정 '+stem)
        
        select1 = random.randint(1,3)
        if select1 == 1:
            ending_list.pop()
            ending_list.append('일정 '+stem+' '+object_do)
        elif select1 == 2:
            ending_list.pop()
            ending_list.append('일정'+' '+object_do+' '+stem)
        elif select1 == 3:
            ending_list.pop()
            ending_list.append(object_do+' '+'일정'+' '+stem)
            
        
    select = random.choice(ref)
    if select:
        ending_list.pop()
        ending_list.append('일정 '+stem)
        select1 = random.choice(ref)
        if select1:
            ending_list.pop()
            ending_list.append('일정 '+stem+' '+object_do)
        
    return ending_list
for i in range(0,20):
    print(few_minute())    
```
